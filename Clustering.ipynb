## Theoretical Questions

1. What is unsupervised learning in the context of machine learning?
  - Unsupervised Learning is a type of machine learning where the model is trained using data without labeled outputs.
  - It automatically discovers hidden patterns, structures, or relationships
within the dataset.
  - Common tasks include clustering, association rule mining,
and dimensionality reduction.


2. How does K-Means clustering algorithm work?
  - K-Means is an unsupervised clustering algorithm that partitions data into K clusters.
  - First, K initial centroids are randomly selected from the dataset.
  - Each data point is assigned to the nearest centroid based on distance (usually Euclidean).
  - The centroids are then recalculated as the mean of all points in each cluster.
  - These steps repeat until the centroids stop changing or convergence is achieved.


3. Explain the concept of a dendrogram in hierarchical clustering.
  - A dendrogram is a tree-like diagram used to visualize the results of hierarchical clustering.
  - It shows how data points or clusters are merged (or split) at different distance levels.
  - The vertical axis represents the distance or dissimilarity between clusters.
  - Clusters that merge at lower heights are more similar than those merged at higher heights.
  - By cutting the dendrogram at a chosen level, we can determine the number of clusters.


4. What is the main difference between K-Means and Hierarchical Clustering?
  - The main difference is that K-Means partitions data into a fixed number (K) of clusters, while Hierarchical Clustering builds a tree-like structure (dendrogram) of clusters.
  - K-Means requires the number of clusters to be specified in advance,
whereas Hierarchical Clustering does not require a predefined number of clusters.
  - K-Means is generally faster for large datasets,
while Hierarchical Clustering is more computationally expensive.


5. What are the advantages of DBSCAN over K-Means?
  - DBSCAN can detect clusters of arbitrary shapes, while K-Means works best with spherical clusters.
  - It does not require specifying the number of clusters in advance.
  - DBSCAN can identify noise and outliers, whereas K-Means assigns every point to a cluster.
  - It works well with clusters of varying densities and is more robust to outliers.


6. When would you use Silhouette Score in clustering?
  - Silhouette Score is used to evaluate the quality of clustering results.
It measures how similar a data point is to its own cluster compared to other clusters.
The score ranges from -1 to 1, where a higher value indicates better clustering.
It is commonly used to determine the optimal number of clusters in algorithms like K-Means.

7. What are the limitations of Hierarchical Clustering?
  - Hierarchical Clustering is computationally expensive and not suitable for very large datasets.
  - Once clusters are merged or split, the process cannot be reversed.
  - It is sensitive to noise and outliers in the data.
  - The choice of distance metric and linkage method can significantly affect the results.


8. Why is feature scaling important in clustering algorithms like K-Means?
  - Feature scaling is important in K-Means because it uses distance measures (like Euclidean distance) to assign data points to clusters. If features have different scales, variables with larger values dominate the distance calculation. This can lead to biased or incorrect cluster formation. Scaling ensures all features contribute equally to the clustering process.


9. How does DBSCAN identify noise points?
  - DBSCAN identifies noise points based on two parameters: epsilon (ε) and minimum points (MinPts). A point is considered a core point if it has at least MinPts within its ε-neighborhood. Points that are not core points and are not reachable from any core point are labeled as noise. These noise points do not belong to any cluster and are treated as outliers.


10.  Define inertia in the context of K-Means.
  - Inertia in K-Means is the sum of squared distances between each data point  and the centroid of its assigned cluster. It measures how tightly the data points are grouped within clusters. Lower inertia indicates more compact and well-formed clusters. It is commonly used to evaluate clustering performance and in the Elbow Method.


11. What is the elbow method in K-Means clustering?
  - The Elbow Method is used to determine the optimal number of clusters (K) in K-Means. It involves plotting the inertia (within-cluster sum of squares) against different values of K. As K increases, inertia decreases, but at some point the rate of decrease sharply changes. The “elbow” point in the graph indicates the optimal K, where adding more clusters gives diminishing returns.


12.  Describe the concept of "density" in DBSCAN.
  - In DBSCAN, density refers to the number of data points within a specified radius (ε) around a point. A region is considered dense if it contains at least a minimum number of points (MinPts). Points in dense regions form clusters, while sparse regions are treated as noise. Thus, clusters are defined as areas of high point density separated by low-density areas.


13. Can hierarchical clustering be used on categorical data?
  - Yes, hierarchical clustering can be used on categorical data, but an appropriate distance/similarity measure must be chosen. Instead of Euclidean distance, measures like Hamming distance or Jaccard similarity are commonly used for categorical variables. The clustering process then builds the dendrogram based on these distances.

14.  What does a negative Silhouette Score indicate?
  - A negative Silhouette Score indicates that a data point is closer to a neighboring cluster than to the cluster it has been assigned to. This suggests possible misclassification or poor cluster separation. It generally means the clustering structure is not well defined.


15. Explain the term "linkage criteria" in hierarchical clustering.
  - Linkage criteria in hierarchical clustering define how the distance between two clusters is calculated during merging. It determines which clusters are combined at each step. Common linkage methods include single linkage (minimum distance), complete linkage (maximum distance), and average linkage (mean distance). The choice of linkage affects the shape and structure of the resulting clusters.


16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
  - K-Means assumes clusters are spherical and of similar size and density. When clusters have different sizes or varying densities, the algorithm may split large clusters or merge smaller dense clusters incorrectly. It also relies on the mean as the centroid, which is sensitive to outliers. As a result, K-Means may produce inaccurate cluster boundaries in such datasets.


17. What are the core parameters in DBSCAN, and how do they influence clustering?
  - The core parameters in DBSCAN are epsilon (ε) and MinPts (minimum points).Epsilon defines the radius of the neighborhood around a data point. MinPts specifies the minimum number of points required within ε to form a dense region. Smaller ε or larger MinPts results in fewer, tighter clusters and more noise, while larger ε or smaller MinPts may merge clusters and reduce noise detection.


18.  How does K-Means++ improve upon standard K-Means initialization?
  - K-Means++ improves standard K-Means by selecting initial centroids more carefully. Instead of choosing centroids randomly, it spreads them out based on distance. Each new centroid is chosen with a probability proportional to its distance from existing centroids. This reduces poor initialization, speeds up convergence, and often results in better and more stable clustering.


19. What is agglomerative clustering?
  - Agglomerative clustering is a type of hierarchical clustering. It starts with each data point as a separate cluster. At each step, the two closest clusters are merged based on a linkage criterion. This process continues until all points are grouped into a single cluster or a desired number of clusters is obtained.


20. What makes Silhouette Score a better metric than just inertia for model evaluation?
  - Silhouette Score evaluates both cluster cohesion and separation, while inertia only measures within-cluster compactness.
  - Inertia always decreases as the number of clusters increases, making it hard to judge the optimal number of clusters.
  - Silhouette Score ranges from -1 to 1 and reflects how well points are assigned.
  - Thus, it provides a more balanced and interpretable measure of clustering quality.

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=400, centers=4, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Display silhouette scores
for k, score in zip(k_values, scores):
    print(f"k = {k}, Silhouette Score = {score:.4f}")

# Plot bar chart
plt.figure()
plt.bar(list(k_values), scores)
plt.title("Silhouette Scores for k = 2 to 5")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.show()
```

31.  Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic overlapping clusters
X, _ = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)

# Create mesh grid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict cluster labels for each point in mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            marker='X')

plt.title("K-Means with Decision Boundaries (Overlapping Clusters)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

```python
# Import required libraries
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
cancer = load_breast_cancer()
X = cancer.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate inertia for K = 2 to 6
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X_scaled)
    inertia_values.append(model.inertia_)

# Print inertia values
for k, inertia in zip(k_values, inertia_values):
    print(f"K = {k}, Inertia = {inertia:.2f}")

# Plot inertia vs K
plt.figure()
plt.plot(list(k_values), inertia_values, marker='o')
plt.title("Inertia vs Number of Clusters (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

```python
# Import required libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)

# Compute silhouette coefficient for each sample
silhouette_vals = silhouette_samples(X_scaled, labels)

# Plot silhouette values for each sample
plt.figure()
plt.scatter(range(len(silhouette_vals)), silhouette_vals)
plt.title("Silhouette Coefficient for Each Sample (Iris + KMeans)")
plt.xlabel("Sample Index")
plt.ylabel("Silhouette Coefficient")
plt.show()

# Print average silhouette score
print("Average Silhouette Score:", round(np.mean(silhouette_vals), 4))
```

47.  Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate noisy blob data
X, _ = make_blobs(n_samples=500, centers=3, cluster_std=1.8, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

# Count clusters (excluding noise label = -1)
unique_labels = set(labels)
num_clusters = len(unique_labels - {-1})

# Count noise points
num_noise = np.sum(labels == -1)

print("Number of clusters (excluding noise):", num_clusters)
print("Number of noise points:", num_noise)
```

48.  Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
# Import required libraries
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load the Digits dataset
digits = load_digits()
X = digits.data

# Use a subset for faster computation
X_subset = X[:800]

# Reduce dimensions using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_reduced = tsne.fit_transform(X_subset)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X_reduced)

# Plot the clusters
plt.figure()
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels)
plt.title("Agglomerative Clustering on Digits (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
